```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df['Amount Scaled']=scaler.fit_transform(df[['Amount']])
df['Time Scaled']=scaler.fit_transform(df[['Time']])
df=df.drop(columns=['Amount','Time','Hour'])
```

```python
X=df.drop(columns='Class')
Y= df['Class']
X.head()
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
from sklearn.model_selection import train_test_split, cross_val_score
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2, random_state=42, stratify=Y)
fraud_count = (Y_train == 1).sum()
k = min(5, fraud_count - 1)
print(f"Using k_neighbors={k}")

X_train_resampled, Y_train_resampled= smote.fit_resample(X_train, Y_train)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, recall_score,f1_score, roc_auc_score,precision_recall_curve, average_precision_score,classification_report
from imblearn.pipeline import Pipeline
#dt_model = DecisionTreeClassifier(random_state=42)
pipeline=Pipeline([('smote',SMOTE(random_state=42)),
                  ('model', DecisionTreeClassifier(random_state=42))])

dt_recall = cross_val_score(pipeline, X_train, Y_train,
                             cv=5, scoring='recall')
dt_f1 = cross_val_score(pipeline, X_train, Y_train,
                         cv=5, scoring='f1')
dt_roc = cross_val_score(pipeline, X_train, Y_train,
                          cv=5, scoring='roc_auc')

print("Decision Tree CV on original data:")
print(f"  Recall  : {dt_recall.mean():.4f} +/- {dt_recall.std():.4f}")
print(f"  F1      : {dt_f1.mean():.4f} +/- {dt_f1.std():.4f}")
print(f"  ROC-AUC : {dt_roc.mean():.4f} +/- {dt_roc.std():.4f}")

pipeline.fit(X_train, Y_train)
val=pipeline.predict(X_test)
val_proba=pipeline.predict_proba(X_test)[:,1]
print('Recall: ',recall_score(Y_test, val))
print('ROC-AUC', roc_auc_score(Y_test, val_proba))
print('F1', f1_score(Y_test, val))


# Fit on SMOTE data — this is where SMOTE helps
#dt_model.fit(X_train_resampled, y_train_resampled)

# Evaluate on REAL test set — never resampled
#y_pred_dt = dt_model.predict(X_test)
#print("\nDecision Tree — Final Test Set:")
#print(f"  Recall  : {recall_score(Y_test, y_pred_dt):.4f}")
#print(f"  F1      : {f1_score(Y_test, y_pred_dt):.4f}")
#print(classification_report(Y_test, y_pred_dt))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, recall_score,f1_score, roc_auc_score,precision_recall_curve, average_precision_score,classification_report
from imblearn.pipeline import Pipeline
rf_model=RandomForestClassifier(random_state=42, n_jobs=-1)
```

```python
from imblearn.pipeline import Pipeline
pipeline_rf=Pipeline([('smote',SMOTE(random_state=42)),
                     ('model',RandomForestClassifier(random_state=42,n_jobs=-1))])


param_grid={'model__n_estimators':[100,200],
            'model__max_depth':[10,20],
            'model__min_samples_leaf':[2,5],
            'model__min_samples_split':[2,5],
            'model__class_weight':['balanced',None]}

grid_search=GridSearchCV(estimator=pipeline_rf,param_grid=param_grid, cv=5, scoring='recall',verbose=1,n_jobs=-1 )
grid_search.fit(X_train, Y_train)
print('best parameters: ', grid_search.best_params_)
print('best score: ',grid_search.best_score_)
```

```python

from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_leaf': [2, 5],
    'min_samples_split': [2, 5],
    'class_weight': ['balanced', None]
}

random_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42, n_jobs=-1),
    param_distributions=param_grid,
    n_iter=10,
    cv=3,
    scoring='recall',
    verbose=1,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train_resampled, Y_train_resampled)
print('Best parameters:', random_search.best_params_)
print('Best score:', random_search.best_score_)
```

```python
pipeline_baseline = Pipeline([
    ('model', RandomForestClassifier(n_estimators=100,
                                      random_state=42, n_jobs=-1))
])

pipeline_weighted = Pipeline([
    ('model', RandomForestClassifier(n_estimators=100,
                                      class_weight='balanced',
                                      random_state=42, n_jobs=-1))
])

pipeline_smote = Pipeline([
    ('smote', SMOTE(random_state=42)),
    ('model', RandomForestClassifier(n_estimators=100,
                                      random_state=42, n_jobs=-1))
])

pipeline_baseline.fit(X_train, Y_train)
pipeline_weighted.fit(X_train, Y_train)
pipeline_smote.fit(X_train, Y_train)

print("All three models trained!")
```

```python
models = {
    'Baseline': pipeline_baseline,
    'Class Weight': pipeline_weighted,
    'SMOTE': pipeline_smote
}

print("=" * 60)
print("RANDOM FOREST — ALL THREE APPROACHES")
print("=" * 60)
print(f"{'Metric':<20} {'Baseline':>12} {'ClassWeight':>12} {'SMOTE':>12}")
print("-" * 60)

results = {}
for name, model in models.items():
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    results[name] = {
        'recall': recall_score(Y_test, y_pred),
        'f1': f1_score(Y_test, y_pred),
        'roc_auc': roc_auc_score(Y_test, y_proba),
        'accuracy': accuracy_score(Y_test, y_pred)
    }

print(f"{'Recall':<20} {results['Baseline']['recall']:>12.4f} {results['Class Weight']['recall']:>12.4f} {results['SMOTE']['recall']:>12.4f}")
print(f"{'F1':<20} {results['Baseline']['f1']:>12.4f} {results['Class Weight']['f1']:>12.4f} {results['SMOTE']['f1']:>12.4f}")
print(f"{'ROC-AUC':<20} {results['Baseline']['roc_auc']:>12.4f} {results['Class Weight']['roc_auc']:>12.4f} {results['SMOTE']['roc_auc']:>12.4f}")
print(f"{'Accuracy':<20} {results['Baseline']['accuracy']:>12.4f} {results['Class Weight']['accuracy']:>12.4f} {results['SMOTE']['accuracy']:>12.4f}")
print("=" * 60)
```

```python
from sklearn.metrics import confusion_matrix
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, (name, model) in zip(axes, models.items()):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(Y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)
    ax.set_title(f'Confusion Matrix — RF {name}')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

plt.tight_layout()
plt.show()
```
